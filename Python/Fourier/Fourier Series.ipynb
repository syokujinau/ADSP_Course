```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
number_of_terms = 10
```

```python
t = np.linspace(-1, 1, 1000)
x = np.zeros(1000)
```

```python
for n in range(1, number_of_terms + 1):
    x += 2 / n * np.pi * (1 - np.power(-1, n)) * np.sin(n * np.pi * t)
```

```python
plt.plot(t, x)
plt.xlabel('t (sec)')
plt.ylabel('Amplitude')
```
