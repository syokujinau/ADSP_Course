```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = 1
T = 1
w = np.linspace(-20, 20, 1000)
X = A * T * np.sinc(w * T / (2 * np.pi))
```

```python
plt.plot(w, X)
plt.xlabel(r'$\omega$')
plt.ylabel(r'$X(\omega)$')
```

# DFT

```python
from numpy.fft import fft, ifft
```

```python
x = np.array([1, 4, 2, 3])
X = fft(x)
```

```python
Xm = abs(X)
xx = ifft(X)
```

```python
print("x = ", x)
print("X = ", X)
print("Magnitude of X =", Xm)
print("Inverse FFT of X = ", xx)
```
